```python
import numpy as np

IGS_data = np.load('IGS_data11_17.npy')
MIT_data = np.load('MIT_data11_17.npy')
TOPEX = np.load('new_topex.npy').reshape((61368, 64, 64, 1))

print(IGS_data.shape)
print(MIT_data.shape)
print(TOPEX.shape)
```

```python
import matplotlib
import matplotlib.pyplot as plt

n = 1000
plt.figure(dpi=300,figsize=(10,4))
plt.subplot(1,3,1)
plt.imshow(IGS_data[n,:,:,0], cmap='jet',vmax=50,vmin=0)

plt.subplot(1,3,2)
plt.imshow(MIT_data[n,:,:,0], cmap='jet',vmax=50,vmin=0)

plt.subplot(1,3,3)
plt.imshow(TOPEX[n,:,:,0], cmap='jet',vmax=50,vmin=0)

plt.show()
```

```python
MIT_miss = MIT_data.copy()
MIT_miss_data = np.ones((61368, 64, 64, 1)) * 9999.
for i in range(len(MIT_miss)):
    MIT_zhi = np.array(np.where(MIT_miss[i] != 9999.))
    MIT_zhi_num = len(MIT_zhi[0])
    MIT_zhi_num_20 = int(MIT_zhi_num*0.15)
    
    import random
    random_num = np.array(random.sample(list(np.arange(0,MIT_zhi_num,1)), MIT_zhi_num_20))
    random_num = random_num .astype('int64')
    one_wei = (MIT_zhi[0])[random_num]
    two_wei = (MIT_zhi[1])[random_num]
    three_wei = (MIT_zhi[2])[random_num]
    
    for j in range(len(one_wei)):
        MIT_miss_data[i,one_wei,two_wei,three_wei] = MIT_data[i,one_wei,two_wei,three_wei]
        MIT_miss[i,one_wei,two_wei,three_wei] = 9999.
    print("\r爬虫01号-第%d次访问！" %i,end= " ")
```

```python
n = 0
plt.figure(dpi=300,figsize=(24,8))
plt.subplot(1,3,1)
plt.imshow(MIT_data[n,:,:,0], cmap='jet',vmax=50,vmin=0)
plt.colorbar()

plt.subplot(1,3,2)
plt.imshow(MIT_miss[n,:,:,0], cmap='jet',vmax=50,vmin=0)
plt.colorbar()

plt.subplot(1,3,3)
plt.imshow(MIT_miss_data[n,:,:,0], cmap='jet',vmax=50,vmin=0)
plt.colorbar()

plt.show()
```

```python
TOPEX_miss = TOPEX.copy()
TOPEX_miss_data = np.ones((61368, 64, 64, 1)) * 9999.
for i in range(len(TOPEX_miss)):
    TOPEX_zhi = np.array(np.where(TOPEX_miss[i] != 9999.))
    TOPEX_zhi_num = len(TOPEX_zhi[0])
    TOPEX_zhi_num_20 = int(TOPEX_zhi_num*0.50)
    
    import random
    random_num = np.array(random.sample(list(np.arange(0,TOPEX_zhi_num,1)), TOPEX_zhi_num_20))
    random_num = random_num .astype('int64')
    one_wei = (TOPEX_zhi[0])[random_num]
    two_wei = (TOPEX_zhi[1])[random_num]
    three_wei = (TOPEX_zhi[2])[random_num]
    
    for j in range(len(one_wei)):
        TOPEX_miss_data[i,one_wei,two_wei,three_wei] = TOPEX[i,one_wei,two_wei,three_wei]
        TOPEX_miss[i,one_wei,two_wei,three_wei] = 9999.
    print("\r爬虫01号-第%d次访问！" %i,end= " ")
```

```python
n = 0
plt.figure(dpi=300,figsize=(24,8))
plt.subplot(1,3,1)
plt.imshow(TOPEX[n,:,:,0], cmap='jet',vmax=50,vmin=0)
plt.colorbar()

plt.subplot(1,3,2)
plt.imshow(TOPEX_miss[n,:,:,0], cmap='jet',vmax=50,vmin=0)
plt.colorbar()

plt.subplot(1,3,3)
plt.imshow(TOPEX_miss_data[n,:,:,0], cmap='jet',vmax=50,vmin=0)
plt.colorbar()

plt.show()
```
